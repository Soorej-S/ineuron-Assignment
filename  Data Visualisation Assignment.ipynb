```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
from sklearn import datasets
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
X = iris.drop(columns = ['species'])
y = iris['species']
pca = PCA(n_components=3)
iris_transform = pca.fit_transform(X)
principal_Df = pd.DataFrame(data = iris_transform, columns = ['principal component 1', 'principal component 2','principal component 3'])
final_df = principal_Df.join(y)
```

```python
fig = plt.figure(1, figsize =(9,9))
ax = Axes3D(fig, elev = -150, azim = 150, auto_add_to_figure=False)
fig.add_axes(ax)
ax.scatter(final_df['principal component 1'],final_df['principal component 2'],final_df['principal component 3'], c = y.replace(iris['species'].unique(),value =(0,1,2)), cmap=plt.cm.rainbow_r)
ax.set_title("Iris Data")
ax.set_xlabel('principal component 1')
ax.set_ylabel('principal component 2')
ax.set_zlabel('principal component 3')
plt.show()  
```
