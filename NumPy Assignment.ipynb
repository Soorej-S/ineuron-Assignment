```python
# 1

def Vandermonde(max_num,mat_size,N):

    arr = np.random.randint(max_num, size=(mat_size))
    return np.column_stack([arr**(N-1-i) for i in range(N)])
    
Vandermonde(5,4,5)
```

```python
# 2

def moving_average(test,n):

    ma_list = []
    for i in range(len(test)-(n-1)):
        var1, var2, var3 = [test[j] for j in range(i,i+n)]
        res = (var1+var2+var3)/n
        ma_list.append(res) 
    print(ma_list)

a = [3, 5, 7, 2, 8, 10, 11, 65, 72, 81, 99, 100, 150]
moving_average(a,3)
```
