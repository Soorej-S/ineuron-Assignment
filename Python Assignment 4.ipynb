```python
#1.1 Write a Python Program(with class concepts) to find the area of the triangle using the below formula.

class length:
    def __init__(self,a,b,c):
        self.a = a
        self.b = b
        self.c = c
        
class triangle(length):
    def area(self):
        s = (self.a + self.b + self.c)/2
        return (s*(s-self.a)*(s-self.b)*(s-self.c)) ** 0.5

A = triangle(5,5,5)
print("Area of triangle : ",A.area())
```

```python
# 1.2 Write a function filter_long_words() that takes a list of words and an integer n and returns the list of words that are longer than n.

def filter_long_words(words,n):
    
    long_words = []

    for i in words.split():
        if len(i)>n:
            long_words.append(i)
        
    print("The list of words greater than integer is", long_words)

filter_long_words("Write a function that takes a list of words and an integer n and returns the list of words that are longer than n.",5)
```

```python
# 2.1 Write a Python program using function concept that maps list of words into a list of integers representing the lengths of the corresponding words.

def word_to_num(words):

    num_list = []

    for i in words.split():
        word_len = len(i)
        num_list.append(word_len)

    print("Mapped list of words into a list of integers is", num_list)

word_to_num("Write a function that takes a list of words and an integer n and returns the list of words that are longer than n.")
```

```python
# 2.2 Write a Python function which takes a character (i.e. a string of length 1) and returns True ifit is a vowel, False otherwise.

def vowels(letter):
    
    for i in ['a','e','i','o','u']:
        if letter == i:
            return True
    else:
        return False
    
print(vowels('u'))
print(vowels('y'))
```
